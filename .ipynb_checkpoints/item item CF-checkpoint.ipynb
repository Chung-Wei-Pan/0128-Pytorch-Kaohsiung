```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ratings = pd.read_csv('booking_ratings.csv')
books = pd.read_csv('books.csv')
users = pd.read_csv('users.csv')
```

```python
print(books.shape)
print(list(books.columns))
```

## Recommendations based on rating counts -- 以評分數為基礎的推薦方法

```python
rating_count = pd.DataFrame(ratings.groupby('ISBN')['bookRating'].count())
rating_count.sort_values('bookRating', ascending=False).head()
```

### 被評分數最高的書籍

```python
most_rated_books = pd.DataFrame(['0971880107', '0316666343', '0385504209', '0060928336', '0312195516'], index=np.arange(5), columns = ['ISBN'])
most_rated_books_summary = pd.merge(most_rated_books, books, on='ISBN')
most_rated_books_summary
```

### 計算每一本書的總評分個數以及平均評分數

```python
average_rating = pd.DataFrame(ratings.groupby('ISBN')['bookRating'].mean())
average_rating['ratingCount'] = pd.DataFrame(ratings.groupby('ISBN')['bookRating'].count())
average_rating.sort_values('ratingCount', ascending=False).head()
```

### 我們將極端值─也就是評分數小於200的users，以及評分數小於100的item去除

```python
counts1 = ratings['userID'].value_counts()
ratings = ratings[ratings['userID'].isin(counts1[counts1 >= 200].index)]
counts = ratings['bookRating'].value_counts()
ratings = ratings[ratings['bookRating'].isin(counts[counts >= 100].index)]
```

### 將所有的評分資料以 One-hot 轉換成矩陣

```python
ratings_pivot = ratings.pivot(index='userID', columns='ISBN').bookRating
userID = ratings_pivot.index
ISBN = ratings_pivot.columns
print(ratings_pivot.shape)
ratings_pivot.head()
```

### 計算 "0316666343" 這本書與其他書籍的相似度
可以發現與該本書相似度最高的書籍約0.47

```python
bones_ratings = ratings_pivot['0316666343']
##Pandas 的 corrwith 計算 Pearson
similar_to_bones = ratings_pivot.corrwith(bones_ratings)
corr_bones = pd.DataFrame(similar_to_bones, columns=['pearsonR'])
corr_bones.dropna(inplace=True)
corr_summary = corr_bones.join(average_rating['ratingCount'])
corr_summary[corr_summary['ratingCount']>=300].sort_values('pearsonR', ascending=False).head()
```

### 取出與 "0316666343" 相似書籍的詳細資料

```python
books_corr_to_bones = pd.DataFrame(['0312291639', '0316601950', '0446610038', '0446672211', '0385265700'], 
                                  index=np.arange(5), columns=['ISBN'])
corr_books = pd.merge(books_corr_to_bones, books, on='ISBN')
corr_books
```
